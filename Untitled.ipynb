```python
# Import dependencies


import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data.csv')
```

```python
plt.plot(data['duration'], data['sstate'], linestyle='none', marker='o', color='b')
plt.title('corelation : duration and sstate')
plt.xlabel('duration')
plt.ylabel('sstate ratio')
plt.show()
```

```python
mst6g = data[data.target == 'mst6g']
plt.subplot(2, 1, 1)                # nrows=2, ncols=1, index=1
plt.plot(mst6g['duration'], mst6g['sstate'],linestyle='none',marker='o', color='r')
plt.title('[mst6g] corelation : duration and sstate')
plt.xlabel('duration')
plt.ylabel('sstate ratio')

gen2clu = data[data.target == 'gen2clu']
plt.subplot(2, 1, 2)                # nrows=2, ncols=1, index=2
plt.plot(gen2clu['duration'], gen2clu['sstate'],linestyle='none',marker='o', color='g')
plt.title('[gen2clu] corelation : duration and sstate')
plt.xlabel('duration')
plt.ylabel('sstate ratio')

plt.tight_layout()
plt.show()
```

```python
plt.hist(data['sstate'])
plt.xlabel('sstate')
plt.ylabel('Count')
plt.title('sstate distribution')
```

```python
bar_data = data['target'].value_counts().reset_index()
bar_data['err'] = data['target'].value_counts().std()
```

```python
plt.figure(figsize=(10,3))
plt.barh(bar_data['index'], bar_data['target'])
plt.xlabel('Ability')
plt.ylabel('Count')
plt.title("target Count")
```
